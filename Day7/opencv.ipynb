# Data Augmentation function

- it will gentare the rotated image in range 0-360
- and filb and brighten random image


```python
import random

def augmented_image(img):
    cv2.imread(img)

    (h, w) = cnic_sign.shape[:2]
    (cX, cY) = (w // 2, h // 2)
    for i in range(1,360,5):
        # rotate our image by 45 degrees around the center of the image
        M = cv2.getRotationMatrix2D((cX, cY), i, 1.0)
        rotated = cv2.warpAffine(cnic_sign, M, (w, h))
        if random.choice([True, False]):
            flipped_image = cv2.flip(rotated, 1)
            brightness_factor = random.uniform(0.5, 2.0)
            brightened_image = cv2.convertScaleAbs(flipped_image, alpha=brightness_factor, beta=0)
        else:
            brightened_image = rotated
        # Save the cropped image
        cv2.imwrite(f"D:\ML1BootCamp\Day7\cnicdataset\cnic_sign{i}.jpg",brightened_image) 
        cv2.waitKey(0)
        cv2.destroyAllWindows()


#function call
path = 'D:\ML1BootCamp\Day7\cnicdataset\cnic_sign.jpg'
augmented_image(path)
```
