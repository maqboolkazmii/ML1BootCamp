# Find columns with high correlation

```python
df.corrwith(df['last_price']).plot()
```

```python
corr_matrix = df1.corr()

correlation_threshold = 0.1


highly_correlated_cols = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > correlation_threshold:
            colname = corr_matrix.columns[i]
            highly_correlated_cols.add(colname)


print("Highly Correlated Columns:", len(highly_correlated_cols))
```

- 149 col have high correlation with target col on correlation threshold 0.1

```python
from statsmodels.graphics.tsaplots import plot_acf

plt.figure(figsize=(12, 6))
plot_acf(df['last_price'], lags=30)
plt.xlabel('Lag')
plt.title('Autocorrelation Plot of Price')
plt.show()
```

- in plot acf the graph is showing data point 0 t0 1 with blue line
- 
The absence of data points towards -1 in the ACF (AutoCorrelation Function) plot suggests a lack of significant negative autocorrelation in the time series data. This indicates that the current observations in the time series are not strongly influenced by the immediately preceding negative observations.

```python
from statsmodels.graphics.tsaplots import plot_pacf

plt.figure(figsize=(12, 6))
plot_pacf(df['last_price'], lags=30)
plt.xlabel('Lag')
plt.title('Partial Autocorrelation Plot of Price')
plt.show()
```

- In partical auto corr graph two values between 0 to 1 and onter values are small change in positive and negative around the 0 line.
- In the partial auto-correlation function (PACF) plot, two values between 0 to 1 usually indicate the significant direct influences of the past observations at those lags on the current observation, while small changes around the 0 line suggest negligible partial correlations with other lags. The PACF helps identify the direct relationship between each lag and the current observation, excluding the indirect effects from shorter lags.

```python
df1['rolling_mean'] = df1['last_price'].rolling(window=3).mean()
df1['rolling_std'] = df1['last_price'].rolling(window=3).std()

```

```python
df1.head(2)
```

```python
plt.figure(figsize=(10, 6))

df1['last_price'].plot(label='Last Price', color='blue')
df1['rolling_mean'].plot(label='Rolling Mean', color='green')
df1['rolling_std'].plot(label='Rolling Standard Deviation', color='red')


plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Stock Data Analysis')
plt.legend()
plt.show()
```

- we have make rolling mean and standard deviation and draw graph
- our Std and rolling mean is constant it means that data is stationary.
- The constancy of the rolling mean and standard deviation indicates that the data is stationary. A stationary time series exhibits consistent statistical properties over time, providing a stable basis for analysis and modeling.

# ADF test

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))


adf_test(df1['last_price'])
```
